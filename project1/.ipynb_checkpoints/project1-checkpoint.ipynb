```python
import pandas as pd
import json
import re #regular expression
```

```python
print("### Structured Data Example ###")

# Structured data (e.g., table from a database)
data = {
    'EmployeeID': [101, 102, 103],
    'Name': ['Ravi', 'Priya', 'Amit'],
    'Department': ['Finance', 'HR', 'IT'],
    'Salary': [65000, 55000, 75000]
}

df = pd.DataFrame(data)
print(df, "\n")

# You can filter, sort, or do SQL-like operations easily
print("Employees with salary > 60000:\n", df[df['Salary'] > 60000], "\n")

```

```python
# Semi-structured data (e.g., JSON, NoSQL)
print("### Semi-Structured Data Example ###")
json_data = '''
[
  {"name": "Ravi", "skills": ["Excel", "SAP"], "details": {"city": "Delhi", "age": 29}},
  {"name": "Priya", "skills": ["Recruitment", "Communication"], "details": {"city": "Mumbai", "age": 31}},
  {"name": "Amit", "skills": ["Python", "AWS"], "details": {"city": "Bangalore", "age": 28}}
]
'''

parsed = json.loads(json_data)
print(json.dumps(parsed, indent=2), "\n")

# Convert semi-structured JSON to structured DataFrame
df_json = pd.json_normalize(parsed)
print("Flattened JSON to Table:\n", df_json, "\n")
```

```python
print("### Unstructured Data Example ###")

# Unstructured data (e.g., text, emails, logs)
text_data = """
Ravi joined finance department last year.
Priya from HR got promoted recently.
Amit is working on cloud migration in IT.
"""

print("Raw text data:\n", text_data)

# Extract useful info using regex or NLP
names = re.findall(r'\b[A-Z][a-z]+\b', text_data)
departments = re.findall(r'Finance|HR|IT', text_data)
print("\nExtracted Names:", names)
print("Extracted Departments:", departments)
```
